## **SCRAPPING DATA**

```python
TARGET_PER_LINK = 25
MIN_WORDS = 40

URLS = {
    "Olahraga Non Sepak Bola": [
        "https://sport.detik.com/raket",
        "https://sport.detik.com/basket",
        "https://sport.detik.com/f1",
        "https://sport.detik.com/moto-gp",
        "https://www.kompas.com/motogp",
        "https://www.kompas.com/badminton",
        "https://www.kompas.com/sports"
    ],
    "Liga Inggris": [
        "https://bola.kompas.com/liga-inggris",
        "https://sport.detik.com/sepakbola/liga-inggris",
        "https://www.bola.net/inggris/"
    ],
    "Liga Indonesia": [
        "https://sport.detik.com/sepakbola/liga-indonesia",
        "https://bola.kompas.com/liga-indonesia",
        "https://www.bola.com/indonesia"
    ],
    "Liga Spanyol": [
        "https://www.bola.com/spanyol",
        "https://sport.detik.com/sepakbola/liga-spanyol",
        "https://www.bola.net/spanyol/"
    ],
    "Liga Italia": [
        "https://sport.detik.com/sepakbola/liga-italia",
        "https://bola.kompas.com/liga-italia",
        "https://www.bola.net/italia/"
    ]
}

options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
options.add_argument("--disable-notifications") # Biar gak muncul popup notif
# options.add_argument("--headless") # Aktifkan ini kalau gak mau liat browsernya 
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
```

```python
final_data = []
visited_links = set()
```

```python
def get_selectors(url):
    # sesuai domain links
    if "detik.com" in url:
        return {
            "list": ["article", "div.media__item", "div.list-content__item"],
            "body": ["div.detail__body-text", "div.itp_bodycontent"],
            "link": "a"
        }
    elif "kompas.com" in url:
        return {
            "list": ["div.article__list", "div.trend__item", "div.articleItem"],
            "body": ["div.read__content", "div.kcm-read-text", "div.clearfix"],
            "link": "a.article__link"
        }
    elif "bola.net" in url:
        return {
            "list": ["div.box-content", "li.list-item"], # Struktur umum bola.net
            "body": ["div.body-detail", "div.article-content"],
            "link": "a"
        }
    elif "bola.com" in url:
        return {
            "list": ["article.articles--iridescent-list-item", "li.articles--iridescent-list-item"],
            "body": ["div.read-page--content-body"],
            "link": "a.articles--iridescent-list-item__link"
        }
    return None
```

```python
def clean_text(soup_element):
    # bersihin noise iklan / script
    for tag in soup_element(["script", "style", "iframe", "a", "strong", "span", "div.related", "h4"]): 
        tag.decompose()
    return soup_element.get_text(" ", strip=True)
```

```python
print("SCRAPPING PAKAI SELENIUM\n")
for label, links in URLS.items():
    print(f"Memproses Kategori: {label}")
    
    for index_url in links:
        print(f"   -> Membuka Indeks: {index_url}")
        
        try:
            driver.get(index_url)
            time.sleep(3) # Tunggu loading awal
            
            for _ in range(4):
                driver.execute_script("window.scrollBy(0, 800);")
                time.sleep(1.5)
            # PARSE HALAMAN INDEKS
            soup_index = BeautifulSoup(driver.page_source, 'html.parser')
            selectors = get_selectors(index_url)   
            
            if not selectors:
                print(f"      [!] Skip: Selector untuk {index_url} belum didefinisikan.")
                continue 
                       
            candidates = []
            for sel in selectors['list']:
                candidates.extend(soup_index.select(sel))
                
            print(f"      Ditemukan {len(candidates)} kandidat link.")
                
            collected_here = 0     

            for card in candidates:
                if collected_here >= TARGET_PER_LINK: 
                    break
                try:
                    
                    link_tag = None
                    
                    if "kompas.com" in index_url:
                        link_tag = card.find("a", class_="article__link")
                        if not link_tag: link_tag = card.find("a", href=True)
                    elif "bola.com" in index_url:
                        link_tag = card.find("a", class_="articles--iridescent-list-item__link")
                        if not link_tag: link_tag = card.find("a", href=True)
                    else:
                        link_tag = card.find("a", href=True)
                        
                    if not link_tag: continue
                    
                    url_artikel = link_tag['href']
                    
                    # Validasi Link
                    if url_artikel in visited_links: continue
                    if any(x in url_artikel for x in ["video", "foto", "photos", "20.detik"]): continue
                    
                    driver.execute_script(f"window.open('{url_artikel}', '_blank');") # Buka tab baru
                    driver.switch_to.window(driver.window_handles[-1]) # Pindah ke tab baru
                    
                    time.sleep(2)
                    
                    soup_art = BeautifulSoup(driver.page_source, 'html.parser')
                    
                    body = None
                    for body_sel in selectors['body']:
                        body = soup_art.select_one(body_sel)
                        if body: break
                    if body:
                        isi_berita = clean_text(body)
                        
                        if len(isi_berita.split()) > MIN_WORDS:
                            final_data.append({
                                "Media": index_url.split('/')[2], # Ambil domainnya aja
                                "Label": label,
                                "Teks": isi_berita,
                                "Link": url_artikel
                            })
                            visited_links.add(url_artikel)
                            collected_here += 1
                            print(f"         [+] ({collected_here}) Berhasil ambil berita.")
                        
                        else:
                            print("         [-] Teks terlalu pendek.")
                    else:
                        print(f"         [-] Gagal ambil body text (Selector beda?).")
                    
                    driver.close()
                    driver.switch_to.window(driver.window_handles[0])
                
                except Exception as e:
                    # Kalau error di tab baru, pastikan tutup dan balik
                    if len(driver.window_handles) > 1:
                        driver.close()
                        driver.switch_to.window(driver.window_handles[0])
                    continue
                
        except Exception as e:
            print(f"      [!] Error di indeks: {e}")
            continue
        
driver.quit()

if final_data:
    df = pd.DataFrame(final_data)
    filename = "Dataset_Berita_Selenium.xlsx"
    df.to_excel(filename, index=False)
    print(f"\nSELESAI! {len(df)} data tersimpan di {filename}")
else:
    print("\n[!] Tidak ada data yang terkumpul.")           
```

```python
def count_custom_entities(text_series, keywords):
    # Gabungkan semua teks menjadi satu string panjang
    all_text = " ".join(text_series.astype(str)).lower()
    
    found_entities = []
    
    for keyword in keywords:
        # Regex \b...\b memastikan kita mencari kata utuh
        # re.escape memastikan karakter spesial di keyword dianggap teks biasa
        pattern = r"\b" + re.escape(keyword) + r"\b"
        
        # Hitung berapa kali keyword muncul
        matches = re.findall(pattern, all_text)
        found_entities.extend(matches)
        
    return Counter(found_entities)
```

```python
unique_labels = df['Label'].unique()

# Loop untuk setiap liga yang ada di data
for label in unique_labels:
    # Cek apakah label ini ada di kamus custom_keywords kita
    # (Misal di data namanya 'Liga Inggris', pastikan key dictionary juga sama)
    if label in custom_keywords:
        print(f"--- Menganalisis Entitas untuk: {label} ---")
        
        # Ambil hanya artikel milik liga tersebut
        subset_text = df[df['Label'] == label]['Teks']
        
        # Ambil daftar keyword khusus liga tersebut
        keywords_target = custom_keywords[label]
        
        # Hitung kemunculan
        entity_counts = count_custom_entities(subset_text, keywords_target)
        
        # Ambil Top 10
        top_10 = entity_counts.most_common(10)
        
        if not top_10:
            print(f"Tidak ada keyword yang ditemukan untuk {label}. Cek kembali daftar keyword/data.")
            continue
            
        # Siapkan data untuk plot
        entities, counts = zip(*top_10)
        
        # Visualisasi Bar Chart
        plt.figure(figsize=(10, 5))
        sns.barplot(x=list(counts), y=list(entities), palette='viridis')
        plt.title(f'Top 10 Custom Entities/Keywords di {label}')
        plt.xlabel('Frekuensi Kemunculan')
        plt.show()
        
        print(f"Detail {label}: {top_10}\n")
        
    else:
        print(f"Peringatan: Label '{label}' tidak ditemukan di custom_keywords. Lewati.")
```

#### **TEXT CLEANING**

```python
def clean_encoding_noise(text: str) -> str:
    if not isinstance(text, str):
        return text

    # 1. Normalisasi berbagai dash ke spasi
    text = re.sub(r"[—–-]", " ", text)   # unicode dash
    text = re.sub(r"-", " ", text)       # hyphen biasa

    # 2. Hapus emoji
    text = re.sub(r"[\U0001F300-\U0001FAFF]", "", text)

    # 3. Hapus karakter encoding rusak
    text = re.sub(r"[âőáÁéÉíÍóÓúÚ]", "", text)

    # 4. Rapikan spasi
    text = re.sub(r"\s+", " ", text).strip()

    return text
```

```python
MODEL_NAME = "indobenchmark/indobert-base-p1"
NUM_LABELS = df["Label"].nunique()

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
```

#### **TOKENISASI**

```python
def tokenize(batch):
    return tokenizer(
        batch["Teks"],
        truncation=True,
        padding="max_length",
        max_length=128
    )

def build_dataset(df):
    ds = Dataset.from_pandas(df[["Teks", "Label"]])
    ds = ds.map(tokenize, batched=True)
    ds = ds.rename_column("Label", "labels")
    ds.set_format("torch", columns=["input_ids", "attention_mask", "labels"])
    return ds

train_ds = build_dataset(train_df)
val_ds   = build_dataset(val_df)
test_ds  = build_dataset(test_df)
```

#### **METRICS EVALUATION**

```python
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=1)

    acc = accuracy_score(labels, preds)
    p, r, f1, _ = precision_recall_fscore_support(
        labels, preds, average="macro", zero_division=0
    )

    return {
        "accuracy": acc,
        "precision_macro": p,
        "recall_macro": r,
        "f1_macro": f1
    }
```

#### **MODEL 1-STAGE WITHOUT HYPERPARAMETER TUNING**

```python
model_1stage = AutoModelForSequenceClassification.from_pretrained(
    MODEL_NAME,
    num_labels=NUM_LABELS
)

args_1stage = TrainingArguments(
    output_dir="./out_1stage",
    eval_strategy="epoch",
    learning_rate=5e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=3,
    logging_strategy="no",
    save_strategy="no",
    report_to="none"
)

trainer_1stage = Trainer(
    model=model_1stage,
    args=args_1stage,
    train_dataset=train_ds,
    eval_dataset=val_ds,
    compute_metrics=compute_metrics
)

trainer_1stage.train()
```

#### **MODEL 2-STAGE**

```python
def build_indobert_mlp(
    model_name,
    num_labels,
    hidden_size=768,
    dropout=0.3
):
    class _IndoBERT_MLP(nn.Module):
        def __init__(self):
            super().__init__()

            self.bert = AutoModel.from_pretrained(model_name)
            self.dropout = nn.Dropout(dropout)
            self.classifier = nn.Sequential(
                nn.Linear(hidden_size, hidden_size),
                nn.ReLU(),
                nn.Linear(hidden_size, num_labels)
            )
            self.loss_fn = nn.CrossEntropyLoss()

        def forward(self, input_ids=None, attention_mask=None, labels=None):
            outputs = self.bert(
                input_ids=input_ids,
                attention_mask=attention_mask
            )

            cls = outputs.last_hidden_state[:, 0, :]
            logits = self.classifier(self.dropout(cls))

            loss = None
            if labels is not None:
                loss = self.loss_fn(logits, labels)

            return SequenceClassifierOutput(
                loss=loss,
                logits=logits
            )

    return _IndoBERT_MLP()
```

```python
args_2stage = TrainingArguments(
    output_dir="./out_2stage",
    eval_strategy="epoch",
    learning_rate=5e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=3,
    logging_strategy="no",
    save_strategy="no",
    report_to="none"
)
```

```python
model_2stage = build_indobert_mlp(
    MODEL_NAME,
    NUM_LABELS
)

trainer_2stage = Trainer(
    model=model_2stage,
    args=args_2stage,
    train_dataset=train_ds,
    eval_dataset=val_ds,
    compute_metrics=compute_metrics
)

trainer_2stage.train()
```

```python
args_2stage_tuned = TrainingArguments(
    output_dir="./out_2stage_tuned",
    eval_strategy="epoch",
    learning_rate=2e-5, # tuned
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=5, # tuned
    logging_strategy="no",
    save_strategy="no",
    report_to="none"
)
```

```python
model_2stage_tuned = build_indobert_mlp(
    MODEL_NAME,
    NUM_LABELS,
    dropout=0.5
)

trainer_2stage_tuned = Trainer(
    model=model_2stage_tuned,
    args=args_2stage_tuned,
    train_dataset=train_ds,
    eval_dataset=val_ds,
    compute_metrics=compute_metrics
)

trainer_2stage_tuned.train()
```

```python
rows = []

def eval_all(trainer, stage, tuning):
    for name, ds in [
        ("Training", train_ds),
        ("Validation", val_ds),
        ("Testing", test_ds)
    ]:
        m = trainer.evaluate(ds)
        rows.append({
            "Model Approach": stage,
            "Hyperparameter Tuning": tuning,
            "Data Split": name,
            "Accuracy": m["eval_accuracy"],
            "Precision (Macro)": m["eval_precision_macro"],
            "Recall (Macro)": m["eval_recall_macro"],
            "F1-Score (Macro)": m["eval_f1_macro"]
        })

eval_all(trainer_1stage, "1-Stage", "Without")
eval_all(trainer_1stage_tuned, "1-Stage", "With")
eval_all(trainer_2stage, "2-Stage", "Without")
eval_all(trainer_2stage_tuned, "2-Stage", "With")

result_table = pd.DataFrame(rows)
result_table
```
